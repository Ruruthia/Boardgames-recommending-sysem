In this notebook we extract names of 20 most popular game mechanics, categories and designers and store them in ../data/LightFM_item_features_names.csv.gz file.
Next, we create ../data/LightFM_item_features.csv.gz file which constains games ids with list of their corresponding [mechanics, categories, designers].

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_json('../data/bgg_GameItem.jl', lines = True)
```

```python
def to_1D(series):
 return pd.Series([x for _list in series for x in _list])

df_mechanics = df.loc[df['mechanic'].notna()]
mechanics = to_1D(df_mechanics['mechanic']).value_counts()
df_categories = df.loc[df['category'].notna()]
categories = to_1D(df_categories['category']).value_counts()
df_designers = df.loc[df['designer'].notna()]
designers = to_1D(df_designers['designer']).value_counts()
```

```python
features = (mechanics.head(20).index).append(categories.head(20).index).append(designers.head(20).index)
features = pd.DataFrame(features)
```

```python
features
```

```python
features.to_csv('../data/LightFM_item_features_names.csv.gz', compression='gzip', index=False)
```

```python
game_features_df = df[['bgg_id' ,'mechanic', 'category', 'designer']].copy()
```

```python
game_features_df
```

```python
game_features_df['features'] = [[] for r in range(len(game_features_df))]                      
```

```python
for index, row in game_features_df.iterrows():
    if type(row['category']) is list:
        row['features'].extend(row['category'])
    if type(row['mechanic']) is list:
        row['features'].extend(row['mechanic'])
    if type(row['designer']) is list:
        row['features'].extend(row['designer'])
```

```python
game_features_df = game_features_df[['bgg_id', 'features']]
game_features_df
```

```python
game_features_df.to_csv('../data/LightFM_item_features.csv.gz', compression='gzip', index=False)
```
